```python
import json
```

```python


with open('prismic.json', encoding='utf-8', errors='ignore') as f:
    prismic = json.load(f)
    
with open('prismic2.json', encoding='utf-8', errors='ignore') as f:
    prismic2 = json.load(f)
```

```python
data = prismic['data']['allSfipages']['edges'] + prismic2['data']['allSfipages']['edges']
```

```python
len(data)
```

```python
cleaner = {x['node']['_meta']['uid']: {'prismic': x['node']['our_take'][0]['text'] for x in data}}
```

```python
import re

def remove_non_alphanumeric(string):
    pattern = r'[^a-zA-Z0-9\s]'
    cleaned_string = re.sub(pattern, '', string).strip()
    return cleaned_string

cleaner = {}
for d in data:
    key = d['node']['_meta']['uid']
    value =  {'prismic': remove_non_alphanumeric(d['node']['our_take'][0]['text'])}

    cleaner[key]= value
```

```python
import csv

with open('SFIs.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        if row != ['brand_tag', 'our_take']:
            try:
                cleaner[row[0]]['django'] = remove_non_alphanumeric(row[1])
            except Exception as e:
                print(e)
                print(row)
                print('----')
```

```python
for k, v in cleaner.items():
    if v.get('prismic') != v.get('django'):
        print(k)
```
